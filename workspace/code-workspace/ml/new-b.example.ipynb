```python
import numpy as np
import pandas as pd
```

```python
transactions = pd.read_csv('data-files/trx_data.csv')
```

```python
transactions.info()
transactions.head()
```

```python
products = transactions['products'].str.split('|')
products
```

```python
transactions['products'] = products.map(lambda x : [ int(v) for v in x])
```

```python
transactions.head()
```

```python
data_list = []
for idx, transaction in transactions.iterrows():
    for product in transaction['products']:
        data_list.append({ 'customerId': transaction['customerId'], 'productId': product })
```

```python
data_list[:5]
```

```python
data = pd.DataFrame(data_list)
data.info()
data.head()
```

```python
data['purchase_count'] = np.ones(len(data))
```

```python
data.head()
```

```python
data = data.groupby(['customerId', 'productId'], as_index=False).count()
```

```python
print( data['purchase_count'].describe() )
data.head()
```
