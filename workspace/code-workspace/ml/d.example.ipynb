```python
# 전체 데이터를 훈련 데이터와 테스트 데이터로 분할

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(diabetes['data'], diabetes['target'], random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
# lr.fit(X_train, y_train) # 데이터로 모델 훈련 (학습)
lr.fit(diabetes['data'], diabetes['target'])
```

```python
print( lr.score(X_train, y_train) )
print( lr.score(X_test, y_test) ) # 모델 평가
```
