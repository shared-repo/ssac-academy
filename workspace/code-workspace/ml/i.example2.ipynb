```python
import pandas as pd
```

```python
message_df = pd.read_csv('data-files/SMSSpamCollection', sep="\t", header=None, names=['label', 'message'])
```

```python
message_df.head()
```

```python
# 직접 반복문 구현
for m in message_df['message'][:10]:
    print(m.lower())
```

```python
# 반복문:       // 외부기능
#   실행문1     // 직접구현기능
#   실행문2     // 직접구현기능

# 외부기능에게 직접구현한 실행문을 함수의 형태로 전달해서 외부기능이 전체작업을 관리하는 동안 필요한 시점에 직접구현한 실행문을 사용 (전달한 함수 호출)
```

```python
def f(r):
    return r.lower()

message_df['message'].map(f) # map 함수는 행을 반복하면서 행을 전달인자로 지정된 함수 f를 호출
```

```python
message_df['message'].map(lambda r: r.lower())
```
