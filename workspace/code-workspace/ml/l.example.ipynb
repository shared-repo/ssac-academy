```python
from sklearn.datasets import load_iris
```

```python
iris_dataset = load_iris()
```

```python
iris_dataset.keys()
```

```python
iris_dataset['target_names'], iris_dataset['target'][:5],  iris_dataset['target'][-5:]
```

```python
import pandas as pd

iris_df = pd.DataFrame(iris_dataset['data'], columns=iris_dataset['feature_names'])
iris_df.info()
iris_df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], 
                                                    random_state=42, test_size=0.2, stratify=iris_dataset['target'])
```

```python
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()
mms.fit(X_train)
X_train_scaled = mms.transform(X_train)
X_test_scaled = mms.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(X_train_scaled, y_train)
```

```python
logreg.score(X_train_scaled, y_train), logreg.score(X_test_scaled, y_test)
```

```python
logreg.predict(X_test_scaled)
logreg.predict([X_test_scaled[0]])
```

```python
# 훈련 및 평가 완료된 변환기, 모델을 파일로 저장

import pickle

with open('data-files/iris-model.pkl', 'wb') as f:
    pickle.dump(logreg, f)

with open('data-files/iris-scaler.pkl', 'wb') as f:
    pickle.dump(mms, f)
```

```python
# 훈련 및 평가 완료된 변환기, 모델을 파일로 저장

import joblib

joblib.dump(logreg, 'data-files/iris-model2.pkl')

joblib.dump(mms, 'data-files/iris-scaler2.pkl')
```

```python
# 저장된 모델을 읽어서 메모리의 객체로 복원
with open('data-files/iris-model.pkl', 'rb') as f:
    logreg2 = pickle.load(f)
```
