```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 데이터 준비
credit_df = pd.read_csv('data-files/credit_train4.csv', encoding='utf-8')
```

```python
credit_df.info()
```

```python
# 저장할 때 index 컬럼을 제거하지 않은 경우 실행
# credit_df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
# 아직 정리되지 않은 Annual Income, Current Loan Amount 컬럼 제거
# credit_df.drop(['Current Loan Amount', 'Annual Income'], axis=1, inplace=True)
```

```python
credit_df.info()
```

```python
# 데이터를 target, features로 분할
X = credit_df.drop("Loan Status", axis=1)
y = credit_df['Loan Status']
```

```python
y.value_counts()
```

```python
# 훈련 데이터와 테스트 데이터로 분할

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
X_train.shape, y_train.value_counts()
```

```python
# 모델 만들기 + 모델 훈련 : KNN
from sklearn.neighbors import KNeighborsClassifier

n = int(X_train.shape[1] / 2)
knn = KNeighborsClassifier(n_neighbors=n)

knn.fit(X_train, y_train)
```

```python
# 모델 평가
knn.score(X_train, y_train), knn.score(X_test, y_test)
```

```python
# 예측
predicted_values = knn.predict(X_test.values[:10])
print(predicted_values)
print(y_test.values[:10])
```

```python
# 모델 만들기 + 모델 훈련 : Logistic Regression

from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
# 평가
logreg.score(X_train, y_train), logreg.score(X_test, y_test)
```

```python
# 예측
predicted_values = logreg.predict(X_test.values[:10])
print(predicted_values)
print(y_test.values[:10])
```

```python
# 모델 만들기 + 모델 훈련 : SVC

from sklearn.svm import SVC

svc = SVC()
svc.fit(X_train, y_train)
```

```python
# 평가
svc.score(X_train, y_train), svc.score(X_test, y_test)
```

```python
# 예측
predicted_values = logreg.predict(X_test.values[:10])
print(predicted_values)
print(y_test.values[:10])
```
