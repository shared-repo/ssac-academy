```python
# 데이터 준비
import statsmodels.api as sm

data = sm.datasets.fair.load_pandas().data # statsmodels 패키지에서 sample data 가져오기
```

```python
data.info()
data.head(5)
```

```python
data['affairs'] = data["affairs"].astype(bool) # 0 -> False /  != 0 -> True
```

```python
data.head()
```

```python
data.groupby(by="affairs")['rate_marriage', 'age', 'yrs_married', 'children', 'religious'].mean()
```

```python
# target과 features 구분
X = data.drop('affairs', axis=1).values
y = data['affairs'].values
```

```python
X[:5]
```

```python
data['affairs'].value_counts()
```

```python
# 훈련 데이터와 테스트 데이터 분할
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
X_train.shape, X_test.shape
```

```python
# 모델 훈련
from sklearn.linear_model import LogisticRegression

lr2 = LogisticRegression() # 2진 분류기 모델 (선형)
lr2.fit(X_train, y_train) # 모델 훈련
```

```python
print( lr2.score(X_train, y_train) )
print( lr2.score(X_test, y_test) )
```
