```python
import numpy as np
import pandas as pd
```

```python
diamonds_df = pd.read_csv('data-files/diamonds.csv')
```

```python
diamonds_df.info()
```

```python
diamonds_df.head()
```

```python
print( diamonds_df.columns[0] )
diamonds_df.drop(diamonds_df.columns[0], axis=1, inplace=True)
diamonds_df.head()
```

```python
diamonds_df2 = pd.get_dummies(diamonds_df)

print(diamonds_df.shape, diamonds_df2.shape)
diamonds_df2.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder

diamonds_df3 = diamonds_df.copy()

color_categories = sorted(diamonds_df3['color'].unique())
print( color_categories )

oe = OrdinalEncoder(categories=[color_categories])
diamonds_df3['encoded_color'] = oe.fit_transform(diamonds_df3["color"].values.reshape(-1, 1))
diamonds_df3.head()
```

```python
diamonds_df3['cut'].unique()
```

```python
y = diamonds_df2['price']
X = diamonds_df2.drop('price', axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()
X_train_scaled = pd.DataFrame(mms.fit_transform(X_train.values), columns=X_train.columns)
X_test_scaled = pd.DataFrame(mms.transform(X_test.values), columns=X_test.columns)
```

```python
X_train_scaled.head()
```

```python
# X_train_scaled[['carat', 'depth']].describe()
X_test_scaled[['carat', 'depth']].describe()
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
y_train.describe()
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso

lr = LinearRegression()
lr.fit(X_train, y_train)
print( lr.score(X_train, y_train), lr.score(X_test, y_test) )
```

```python
train_predicted_values = lr.predict(X_train)
test_predicted_values = lr.predict(X_test)
print( mean_absolute_error(y_train, train_predicted_values), mean_absolute_error(y_test, test_predicted_values) )
print( np.sqrt(mean_squared_error(y_train, train_predicted_values)), np.sqrt(mean_squared_error(y_test, test_predicted_values)) )
```

```python
ridge = Ridge(alpha=0.0001)
ridge.fit(X_train, y_train)
print( ridge.score(X_train, y_train), ridge.score(X_test, y_test) )
```

```python
lasso = Lasso(alpha=0.0001)
lasso.fit(X_train, y_train)
print( lasso.score(X_train, y_train), lasso.score(X_test, y_test) )
```

```python
from sklearn.tree import DecisionTreeRegressor

dtr = DecisionTreeRegressor(max_depth=12)
dtr.fit(X_train, y_train)
print( dtr.score(X_train, y_train), dtr.score(X_test, y_test) )
```

```python
from sklearn.model_selection import GridSearchCV

params = { 'max_depth': [8, 10, 12, 14, 16, 18] }
dtr2 = DecisionTreeRegressor()

grid_cv = GridSearchCV(dtr2, param_grid=params, cv=5)

grid_cv.fit(X_train, y_train)
```

```python
grid_cv.best_params_
```

```python
grid_cv.score(X_train, y_train), grid_cv.score(X_test, y_test)
```

```python
test_predicted_values = grid_cv.predict(X_test)

mean_absolute_error(y_test, test_predicted_values)
```
