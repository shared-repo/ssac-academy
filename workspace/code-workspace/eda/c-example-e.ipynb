```python
recruit_items = soup.select('#recruit_info_list .item_recruit')
# len( recruit_items )

for recruit_item in recruit_items:
    cname = recruit_item.select_one('.area_corp a')
    title = recruit_item.select_one('.area_job .job_tit')
    exp = recruit_item.select_one('.area_job .job_condition span:nth-child(2)')
    edu = recruit_item.select_one('.area_job .job_condition span:nth-child(3)')
    loc = recruit_item.select_one('.area_job .job_condition span:nth-child(1)')
    print("[", cname.get_text().strip(), title.get_text().strip(), exp.get_text().strip(), edu.get_text().strip(), loc.get_text().strip(), "]")
```

```python
search_word = '하둡'
page_no = 1

rows = []

url = "https://www.saramin.co.kr/zf_user/search/recruit?search_area=main&search_done=y&search_optional_item=n&searchType=search&searchword={0}&recruitPage={1}&recruitSort=relation&recruitPageCount=40&inner_com_type=&company_cd=0,1,2,3,4,5,6,7,9,10&quick_apply=&except_read="

while True:
    
    resp = requests.get(url.format(search_word, page_no))

    soup = BeautifulSoup(resp.text, "html.parser")

    recruit_items = soup.select('#recruit_info_list .item_recruit')

    if recruit_items == None or len(recruit_items) == 0:
        print("End of List")
        break

    for recruit_item in recruit_items:
        cname = recruit_item.select_one('.area_corp a')
        title = recruit_item.select_one('.area_job .job_tit')
        exp = recruit_item.select_one('.area_job .job_condition span:nth-child(2)')
        edu = recruit_item.select_one('.area_job .job_condition span:nth-child(3)')
        loc = recruit_item.select_one('.area_job .job_condition span:nth-child(1)')
        rows.append( [ cname.get_text().strip(), title.get_text().strip(), exp.get_text().strip(), edu.get_text().strip(), loc.get_text().strip() ] )

    page_no += 1

    time.sleep(1)
```

```python
print( len(rows) )
rows[:3]
```

```python
recruit_df = pd.DataFrame(data=rows, columns=['cname', 'title', 'exp', 'edu', 'loc'])
recruit_df.head(3)
```

```python
recruit_df.to_csv('data-files/saramin-result.csv', index=False)
```
