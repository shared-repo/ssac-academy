```python
# 25번 상품을 두 개 이상 구매하면 5% 청구 할인
```

```python
def discount(row):
    if int(row[3]) == 25 and int(row[4]) > 1:
        row[5] = str(float(row[5]) * 0.95)
    return row

transaction_data2 = transaction_data2.mapValues(discount)
```

```python
transaction_data2.filter(lambda row: int(row[1][3]) == 25 and int(row[1][4]) > 1).take(5)
```

```python
# 구매액이 가장 큰 고객 찾기
```

```python
amounts = transaction_data2.mapValues(lambda row: float(row[5]))
```

```python
amounts.take(5)
```

```python
totals = amounts.foldByKey(0, lambda x1, x2: x1 + x2)
```

```python
totals.take(5)
```

```python
totals.sortBy(lambda x: -x[1]).take(5)
```

```python
# 81번 상품을 5개 이상 구매한 고객에게 70번 상품을 사은품으로 제공
```

```python
import copy

def add_gift(row):
    if int(row[3]) == 81 and int(row[4]) >= 5:
        gift = copy.copy(row)
        gift[3] = '70'
        gift[4] = '1'
        gift[5] = "0.00"
        
        return [row, gift]
    else:
        return [row]
    
transaction_data2 = transaction_data2.flatMapValues(add_gift)
```

```python
transaction_data2.filter(lambda row: int(row[1][3]) == 70).take(3)
```

```python
transaction_data2.sortBy(lambda row: -float(row[1][5])).take(10)
```

```python
# 최대 구매 고객과 최다 구매 고객에게 사은품 증정 (4번 상품, 93번 상품)
```

```python
gift_transaction = [['2015-03-30', '23:59 PM', '53', '4', '1', '0.00'],['2015-03-30', '23:59 PM', '76', '93', '1', '0.00']]
gift_rdd = sc.parallelize(gift_transaction)
gift_rdd = gift_rdd.map(lambda row: (int(row[2]), row))
```

```python
gift_rdd.collect()
```

```python
transaction_data2 = transaction_data2.union(gift_rdd)
```

```python
# 처리된 데이터를 HDFS의 파일로 저장
```

```python
tmp = transaction_data2.map(lambda row: '#'.join(row[1]))
```

```python
tmp.take(5)
```

```python
tmp.saveAsTextFile('data-files/processed_transaction')
```

```python
# 상품을 Key로 데이터 재구성

# processed_transaction.map(lambda line: line.split('#')).take(5)

# processed_transaction.map(lambda line: line.split('#')) \
#                      .map(lambda row: (int(row[3]), row)).take(5)

transaction_data3 = processed_transaction.map(lambda line: line.split('#')) \
                                         .map(lambda row: (int(row[3]), row))
```

```python
transaction_data3.take(5)
```

```python
# transaction_data3.mapValues(lambda v: float(v[5])).take(5)

# transaction_data3.mapValues(lambda v: float(v[5])) \
#                  .reduceByKey(lambda v1, v2: v1 + v2).take(5)

total_by_product = transaction_data3.mapValues(lambda v: float(v[5])) \
                                    .reduceByKey(lambda v1, v2: v1 + v2) # 1, 2, 3, 4 -> 1 + 2 .. 3 + 3 .. 6 + 4
```

```python
total_by_product.take(5)
```
